```python
import matplotlib.pyplot as plt
import networkx as nx

import matplotlib.pyplot as plt
import networkx as nx

G=nx.Graph()

#Adding nodes
G.add_node("9:D")
G.add_node("5:D")
G.add_node("3:D")
G.add_node("1:D")
G.add_node("4:D")
G.add_node("6:D")
G.add_node("14:D")
G.add_node("11:D")
G.add_node("10:D")
G.add_node("12:D")
G.add_node("17:D")
G.add_node("15:D")
G.add_node("19:D")

#adding edges
G.add_edge("9:D", "5:D")
G.add_edge("5:D", "3:D")
G.add_edge("5:D", "6:D")
G.add_edge("3:D", "1:D")
G.add_edge("3:D", "4:D")
G.add_edge("9:D", "14:D")
G.add_edge("14:D", "11:D")
G.add_edge("11:D", "10:D")
G.add_edge("11:D", "12:D")
G.add_edge("14:D", "17:D")
G.add_edge("17:D", "15:D")
G.add_edge("17:D", "19:D")


#Specifying the positions of the nodes
pos={
    "9:D":(4,20),
    "5:D":(-50,12),
    "3:D":(-100,4),
    "1:D":(-150,-4),
    "4:D":(-80,-4),
    "6:D":(-8,4),
    "14:D":(50,12),
    "11:D":(50,4),
    "10:D":(5,-5),
    "12:D":(100,-5),
    "17:D":(150,5),
    "15:D":(160,-4),
    "19:D":(250,-4)    
}


#Display how the nodes should look
nx.draw(G,pos=pos,with_labels=True,
        node_color="red",node_size=1300,
        font_color="gold",font_size=20,font_family="Times New Roman",
        edge_color="black",
        width=5)


plt.margins(0.2)
plt.show()

```
